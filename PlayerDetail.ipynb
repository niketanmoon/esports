```python
players_dataset = pd.read_csv("csv_files/Player Data 20220928.csv")
```

```python
list_players = players_dataset.playername.unique().tolist()
list_players.remove("unknown player")
```

```python
def averagePlayersData(player_df, name):
    player_dict = {
        "name": name
    }
    dataset = player_df[player_df.playername == name]
    
    total_games_played = len(dataset)
    total_win = len(dataset[dataset.result==1])
    total_loss = len(dataset[dataset.result==0])
    assists = dataset.assists.sum()
    deaths = dataset.deaths.sum()
    kills = dataset.kills.sum()
    KDA = f"{kills}/{deaths}/{assists}"
    
    red_side = len(dataset[dataset.side == "Red"])
    blue_side = len(dataset[dataset.side == "Blue"])
    
    player_dict.update({
        "games_played": total_games_played,
        "total_wins": total_win,
        "total_losses": total_loss,
        "assists": assists,
        "deaths": deaths,
        "kills": kills,
        "KDA": KDA,
        "red_side": red_side,
        "blue_side": blue_side
    })
    
    return pd.DataFrame(data=[player_dict])
```

```python
final_df = pd.DataFrame()
# df = averagePlayersData(player_data, "Soboro")
# final_df = final_df.append(df)
```

```python
for p_name in list_players:
    df = averagePlayersData(players_dataset, p_name)
    final_df = final_df.append(df)
```

```python
final_df.reset_index(drop=True, inplace=True)
```
